### convert word list into dictionary of letters

```python
def convert(wl):
    wd = {}
    for i in wl:
        tp = tuple(i+' ')
        if tp in wd:
            wd[tp]+=1
        else:
            wd[tp] = 1
    return wd
```

```python
from collections import Counter

def countpairs(wd):
    count_pair = Counter()
    for k,v in wd.items():
        if len(k)==1:
            return("terminate")
        for i in range(len(k)-1):
            pair = (k[i],k[i+1])
            count_pair[pair]+=v
    freqpair = max(count_pair,key = lambda x : count_pair[x])
    return freqpair
```

```python
fp = countpairs(wd)
print(fp)
```

```python
def mergepairs(wd,freqpair):
    update = {}
    for k,v in wd.items():
        new=[]
        i=0
        while i<len(k):
            if i < len(k)-1 and freqpair==(k[i],k[i+1]):
                new.append(''.join(freqpair))
                i+=2
            else:
                new.append(k[i])
                i+=1
        update[tuple(new)]=v
    return update
```

```python
mergepairs(wd,fp)
```

```python
wd = convert(wl)
while True:
    fp = countpairs(wd)
    if fp == "terminate":
        break
    wd = mergepairs(wd,fp)
```

### Better method:

```python
def get_counts(token):
    counts = {}
    for pair in zip(token,token[1:]):  ##pythonic way of iterating over pairs
        counts[pair] = counts.get(pair,0)+1  #if pair exist, then increase count by 1. if not, then assign 0 and increment by 1.
    return(counts)

print(get_counts([2,3,4,2,3,5,6,4,3,2,3,4,2,5,6,3]))
```

```python
# sort: find most common pair
pairs = get_counts(enc)
print(sorted(((v,k) for k,v in pairs.items()),reverse=True)) 
#if v is first value in tuple, sorted automatically sorts using value(which is the count of the pair)

#OR
print("------------------------")

top_pair = max(pairs,key=pairs.get)
print(top_pair)
```

### Merge
* Now we want to swap every occurence of this most common pair with a new token (once choice could be be 256, because currently the tokes go from 0-255 so if we are creating a new token we use 256)

```python
def merge(tokens,pair,newtoken):
    i=0
    newtokens = []
    while i<len(tokens):
        if i<len(tokens)-1 and pair[0] == tokens[i] and pair[1]==tokens[i+1]:
            newtokens.append(newtoken)
            i+=2
        else:
            newtokens.append(tokens[i])
            i+=1
    return newtokens

print(merge([5,6,6,7,9,1],(6,7),99))
```

```python
txt = "We begin our song with the Heliconian Muses who hold high and holy mount Helicon and with their soft feet dance upon the violet waters and the altar of Cronos’ mighty son. After bathing their tender flesh in the Permessus, Hippocrene, or sacred Olmedius, they perform their chorals on the highest peaks of Helicon, sublime beauty flowing freely from their feet. Starting there and shrouded in a great mist, they travel by night, their beautiful voices singing to aegis-bearing Zeus, to queen Hera of Argos who walks with golden sandals, to Athena, gleaming-eyed daughter of Zeus, to Phoebus Apollo, to shedder of arrows Artemis, to Earth-holder and Earth-shaker Poseidon, to venerable Themis, to darting-eyed Aphrodite, to golden-crowned Hebe, to fair Dione, to Leto and Iapetus and wily Cronos, to Eos and great Helios and bright Selene,to Gaia and great Oceanus and black Nyx, and to all other immortals who always are.1 The Muses once taught Hesiod a beautiful song as he tended lambs under holy Helicon. First, the goddesses, Olympian Muses, daughters of aegis-bearing Zeus, spoke these words to me: “Shepherds in fields, foul disgraces, nothing but bellies—we know how to make lies into truth, and we know how to speak truth when we wish.” So saying, Zeus’ eloquent daughters plucked a staff, a fine branch of blooming laurels, and gave it to me; they blew a divine voice into my mouth so I would glorify what was and what will be;"
```

```python
tokens = list(map(int,txt.encode("utf=8")))

print(len(tokens))
```

```python
merges = {}

vocab_size = 276

newtokens = list(tokens)

for i in range(vocab_size-256):
    pairs = get_counts(newtokens)
    top_pair = max(pairs, key=pairs.get)
    idx = i+256
    newtokens = merge(newtokens,top_pair,idx)
    print(f"Merge{top_pair} into a new token {idx}")
    merges[top_pair]= idx
```
