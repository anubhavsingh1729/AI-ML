```python
text = "From the Ελικωνικές Μούσες let us begin to sing, who hold the great and holy mount of Ελικών, and dance on soft feet about the deep-blue spring and the altar of the almighty son of Cronos, and, when they have washed their tender bodies in Permessus or in the Horse's Spring or Olmeius, make their fair, lovely dances upon highest Helicon and move with vigorous feet."

token = text.encode("utf-8")
```

```python
token = list(map(int,token))
```

```python
print(text)
print("length: ", len(text))
print("----------------")
print(token)
print("length: ", len(token))
```

```python
vocab = []
d={}
for i in range(len(text)-1):
    pair = "".join([text[i],text[i+1]])
    if pair in d:
        d[pair] += 1
    else:
        d[pair] = 0
```

```python
print(d)
```

```python
for i in range(6):
    mp = max(d, key = lambda k:d[k])
    vocab.append(mp)
```
